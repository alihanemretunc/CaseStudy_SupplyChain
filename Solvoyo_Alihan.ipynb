```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('CustomerSuccessConsultantCaseStudy-C.xlsx')
```

```python
df['Shortfall'] = df['Target Safety Stock'] - (df['Starting Inventory'] - df['Demand'])
```

```python
df['MOQ Constraint'] = df['Shortfall'] >= df['Minimum Order Quantity']
```

```python
df.loc[df['Starting Inventory'] - df['Demand'] >= df['Target Safety Stock'], 'Final Check'] = 0
df.loc[(df['Shortfall'] >= df['Minimum Order Quantity']) & (df['Shortfall'] % df['Incremental Order Quantity'] == 0), 'Final Check'] = df['Shortfall']
df.loc[(df['Shortfall'] >= df['Minimum Order Quantity']) & (df['Shortfall'] % df['Incremental Order Quantity'] != 0), 'Final Check'] = df['Shortfall'] + (df['Incremental Order Quantity'] - (df['Shortfall'] % df['Incremental Order Quantity']))
df.loc[(0 < df['Shortfall']) & (df['Shortfall'] < df['Minimum Order Quantity']), 'Final Check'] = df['Minimum Order Quantity']
df.loc[df['Shortfall'] <= 0, 'Final Check'] = 0
```
